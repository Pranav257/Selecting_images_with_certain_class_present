```python
#!/usr/bin/env python
# coding: utf-8

# In[3]:


import tensorflow as tf
import tensorflow_hub as tfhub
import matplotlib.pyplot as plt

# Pretrained model
model = tfhub.load('https://tfhub.dev/google/HRNet/coco-hrnetv2-w48/1')

# Folder path containing the images
folder_path = 'sub_sel/'

# Get a list of file paths in the folder
file_paths = tf.data.Dataset.list_files(folder_path + '*.jpeg')

# Read and decode each image file
dataset = file_paths.map(lambda x: (x, tf.io.decode_jpeg(tf.io.read_file(x), channels=3)))

# Counter for limiting the output
counter = 0

# Iterate over the dataset and process the images
for file_path, image in dataset:
    # Print the file name
    print("File Name:", file_path.numpy().decode())

    # Preprocess the image
    processed_image = tf.cast(image, tf.float32) / 255.0

    # Run the pretrained model on the image
    resp = model.predict([processed_image])
    print("Output shape:", resp.shape)

    # Get the main classes it thinks it sees
    myIm = resp.numpy()
    ax_values = [17, 18]  # Update to only include ax values of interest
    display_image = False  # Flag to determine if image should be displayed
    for ax in ax_values:
        test = myIm[:, :, :, ax].max()
        if test > 0.5:
            print("Ax:", ax, "Prediction:", test)
            display_image = True
            break  # Exit the loop if ax value is found

    if display_image:
        # Display the image using Matplotlib
        plt.imshow(image)
        plt.axis('off')  # Disable axis ticks and labels
        plt.show()

        # Increment the counter
        counter += 1

    # Break the loop if the counter reaches five
    if counter == 5:
        break


# In[ ]:
```
